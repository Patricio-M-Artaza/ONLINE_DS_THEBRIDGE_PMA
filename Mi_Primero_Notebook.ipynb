# Operaciones Booleanas

Vamos a empezar a concatenar comparaciones con la **Algebra de Bool**. Las operaciones mas comunes seran: *AND*, *OR* Y *NOT*.  
**AND** > Se tienen que cumplir ambas condiciones para que sea `True`  
**OR** > Basta que se cumpla al menos una condicion para que sea `True`  
**NOT** > Lo contrario de lo que haya.

````python
# Ram de 16, 32 o 64 - Procesador i3 con 500gb de disco - Que no pase los 800 de precio.

# Primer Ordenador
ram1 = 32
proce1 = "i5"
disco1 = 500
precio1 = 850

# Segundo Ordenador

ram2 = 8
proce2 = "i5"
disco2 = 500
precio2 = 600

# Tercer Ordenador

ram3 = 32
proce3 = "i3"
disco3 = 500
precio3 = 780
````

````python
# Calculamos el valor de las condiciones por separado para el primer ordenador
cond_ram1 = (ram1 == 16 or ram1 == 32 or ram1 == 64)
cond_proce1 = (proce1 == "i3" and disco1 == 500)
cond_precio1 = (precio1 < 800)
cond_total1 = cond_ram1 and cond_precio1 and cond_proce1
print(cond_total1, cond_precio1, cond_proce1, cond_ram1)
````

````python
# Calculamos el valor de las condiciones por separado para el segundo ordenador
cond_ram2 = (ram2 == 16 or ram2 == 32 or ram2 == 64)
cond_proce2 = (proce2 == "i3" and disco2 == 500)
cond_precio2 = (precio2 < 800)
cond_total2 = cond_ram2 and cond_precio2 and cond_proce2
print(cond_total2, cond_precio2, cond_proce2, cond_ram2)
````

````python
# Calculamos el valor de las condiciones por separado para el Tercer ordenador
cond_ram3 = (ram3 == 16 or ram3 == 32 or ram3 == 64)
cond_proce3 = (proce3 == "i3" and disco3 == 500)
cond_precio3 = (precio3 < 800)
cond_total3 = cond_ram3 and cond_precio3 and cond_proce3
print(cond_total3, cond_precio3, cond_proce3, cond_ram3)
````

# Metodos

````python
# String  
ejemplo = "string en minusculas"

# Metodo para poner un string TODO EN MAYUSCULAS > upper()

print(ejemplo.upper())

# Todo en MINUSCULAS > lower()
print("ESTO ES UN METODO DIFERENTE".lower())
ejemplo1 = "ESTO ES UN METODO DIFERENTE"
print(ejemplo1.lower())

# Metodo para reemplazar uno o varios caracteres > replace()
print(ejemplo.replace("u","a")) # reemplaza una por otra
print(ejemplo.replace("minusculas","otro ejemplo")) # reemplaza palabras enteras
print(ejemplo.replace("minusculas","")) # Reemplaza palabras por espacios  

# String a una lista que es una coleccion de objetos ordenados > split()
print("Esto se va a convertir en una lista de string".split(" ")) # En este caso hace una lista por cada palabra.

# Conocer la posicion dentro de un String > index()
print(ejemplo.index("m")) # El conteo empieza SIEMPRE en 0.

````

## if/elif/else  
Son condicionales para tomar decisiones.

# Diccionarios  
## Sirven para agrupar datos y establecer una relacion entre ambos con la forma clave:valor.  
*La sintaxis es con `{}`*  
**nombre_diccionario = {clave1:valor1, clave2:valor2, clave3:valor3}**  
**Los diccionarios son mutables y desordenados porque su clave esta, justamente, en clave:valor**

````python
# Diccionario con elementos de un coche
coche_dict = {
    "color": "Rojo",
    "marca": "Tesla",
    "modelo": "Cybertruck",
    "peso": 3500,
    "cambio automatico": True  # Color, marca... son las CLAVES.
}
coche_dict
````

````python
# Accesos, modificacion, Aniadir, eliminar elementos en un diccionario.  

# Acceso
print(coche_dict["color"]) 
print(coche_dict["peso"]) # Se puede acceder en cualquier orden.

# Modificacion
coche_dict["color"] = ["Rojo", "gris"] # De esta forma se cambia el color, por ejemplo.
print(coche_dict)

# Aniadir
coche_dict["puertas"] = 6
print(coche_dict)

# Eliminar un elemento
del coche_dict["puertas"]
print(coche_dict)

# Conocer las claves del diccionario
print(coche_dict.keys())

# Valores del diccionario
print(coche_dict.values())
````

````python
# Crear un diccionario vacio  
tienda = {} # diccionario vacio
tienda["nombre"] = "Ruperta" # Aca vamos agregando los datos al diccionario
print(tienda)
````

````python
# Errores en los diccionarios
#coche_dict["atomizadores"] #KeyError cuando se intenta acceder a una clave que no existe.

# Solucion: Se puede usar el bloque try/except o mi_diccionario.get(clave, valor si no existe)
atomiza = coche_dict.get("atomizadores", "no") # Aca pondra el valor de no 
print(atomiza)
marca = coche_dict.get("marca", "desconocido") # Aca como si esta la marca pondra la marca directamente.
print(marca)
````

````python
# Iterar en un diccionario (manera menos elegante)

for i in coche_dict:
    print(i)
    print(coche_dict[i])
````

````python
# Manera mas elegante de iterar con los diccionarios

for i in coche_dict:
    print(i,":", (coche_dict[i]))
````

````python
coche_dict.items() # Esto devuelve una serie de tuplas
````

````python
# Otra manera, muy usada  

for clave,valor in coche_dict.items():
    print(clave,":", valor)
````

## Argumentos posicionales  
**Son aquellos que la funcion distingue por su posicion. Hay que tener en cuenta dos cosas: EL ORDEN en el cual llamamos a la funcion  
tiene que ser el mismo que esta DECLARADO en la funcion y por otro lado son OBLIGATORIOS, es decir, tenemos que llamar a TODOS los
argumentos**

````python
def multiplica(x1, x2, x3, x4):
    resultado = (x1 * x2 * x3) / x4
    return resultado
multiplica(3, 4, 5, 6) # Como es ultima linea devolera el resultado sin el print
````

````python
multiplica(x1 = 4, x4 = 3, x2 = 5, x3 = 6) # Aca estamos cambiando el valor de los argumentos, por ende el resultado sera distinto.
````
