# Conversion de tipos

Se pueden realizar la conversion de tipos de datos, como por ejemplo, pasar de int a str y viceversa. Hay que tener ***CUIDADO*** con lo que hacemos ya que podemos perder info o dar error.

````python
num_real = 24.69
print(type(num_real))

num_entero = int(num_real)
print(num_entero)
print(type(num_entero)) # Al pasar de float a int perdemos unos decimales que NO ES GRAVE aunque va a depender de la apliacion.
````

````python
# FUNCION ROUND: Sirve para redondear un float y para decidir que cantidad de decimales queremos.
num_real = 24.686
num_redondo = round(num_real, 2) # El numero 2 significa la cantidad de decimales que quiero que tenga el redondeo.
print(num_redondo)
````

````python
num_real = 24.65
num_redondo = round(num_real, 0)
print(num_redondo, type(num_redondo))
print(int(num_redondo)) # Aca seria un entero
````

````python
# PASAR DE NUMERO A STRING 
num_real = 24.69
num_entero = 12
real_str = str(num_real)
entero_str = str(num_entero)
print(real_str, "su tipo", type(real_str))
print(entero_str, "su tipo", type(entero_str))
````

````python
print(num_real + num_entero)
print(real_str + entero_str) # Aca concatena agregando el 12 y nada mas ya que es un str.
````

````python
# STRING A NUMERO
mi_entero = "98"
mi_cadena = int(mi_entero)
print(mi_cadena, "el tipo es", type(mi_cadena))
print(mi_entero, "el tipo es", type(mi_entero))

mi_cadena_real = "98.89"
mi_real = float(mi_cadena_real)
print(mi_real, "tipo", type(mi_real))
````
