## Ejercicio 1
Para este ejercicio vamos a poner en práctica [las funciones built in](https://docs.python.org/3/library/functions.html).

1. Calcula el máximo de la lista: [4, 6, 8, -1]
2. Suma todos los elementos de la lista anterior
3. Redondea este float a 3 dígitos decimales: 63.451256965
4. Valor absoluto de: -74

```python
# Max de la lista
lista = [4, 6, 8, -1]
numero_mayor = max(lista)
print(numero_mayor)

# Suma de todos los elementos
suma = sum(lista)
print(suma)

# Redondeo de float
numero = 63.451256965
redondeo = round(numero, 3)
print(redondeo)

# Valor absoluto
number = -74
absoluto = abs(number)
print(absoluto)
```

## Ejercicio 2
Para el siguiente string se pide imprimir por pantalla los siguientes casos

    "En un lugar de la Mancha, de cuyo nombre no quiero acordarme."

1. Pásalo todo a mayúsculas
2. Pásalo todo a minúsculas
3. Solo la primera letra de cada palabra en mayúscula, el resto en minúscula
4. Crea una lista dividiéndolo por sus espacios
5. Sustituye las comas `,` por puntos y comas `;`
6. Elimina las `a` minúsculas

```python
# Mayusculas
print("En un lugar de la Mancha, de cuyo nombre no quiero acordarme.".upper())

# Minusculas
print("En un lugar de la Mancha, de cuyo nombre no quiero acordarme.".lower())

# Primera mayuscula, resto minuscula
print("En un lugar de la Mancha, de cuyo nombre no quiero acordarme.".title())

# Lista
frase = "En un lugar de la Mancha, de cuyo nombre no quiero acordarme."
frase_final = frase.split()
print(frase_final)

# Sustitucion
print("En un lugar de la Mancha, de cuyo nombre no quiero acordarme.".replace(",", ";"))

# Elimina las a minusculas
print("En un lugar de la Mancha, de cuyo nombre no quiero acordarme.".replace("a", ""))
```

## Ejercicio 3
1. Crea una lista con 3 elementos numéricos
2. Añade un cuarto elemento
3. Calcula la suma de todos
4. Elimina el segundo elemento de la lista
5. Añade otro elemento en la posicion 3 de la lista
6. Crea otra lista con 4 elementos y concaténala a la que ya tenías.
7. Ordena la lista de menor a mayor.
8. BONUS: Ordena la lista de mayor a menor.  
(*PISTA: Para hacer 8, tendrás que buscar en la información del método...*)  
  
  
NOTA: Pinta por pantalla el resultado de cada paso.

```python
# Lista
mi_lista = [4, 8, 10]

# Aniadir cuarto elemento
mi_lista.append(11)
print(mi_lista)

# Suma
suma = sum(mi_lista)
print(suma)

# Eliminar el segundo elemento
mi_lista.pop(2)
print(mi_lista)

# Aniadir otro elemento
mi_lista.insert(3, 19)
print(mi_lista)

# Lista nueva
lista_nueva = [11, 33, 56, 90]
lista_concatenada = lista_nueva + mi_lista
print(lista_concatenada)

# Ordenar la lista
lista_concatenada.sort()
print(lista_concatenada)

# Mayor a menor
lista_concatenada.sort(reverse=True)
print(lista_concatenada)
```
